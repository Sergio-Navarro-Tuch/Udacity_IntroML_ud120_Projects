```python
#!/usr/bin/python

""" 
    Starter code for exploring the Enron dataset (emails + finances);
    loads up the dataset (pickled dict of dicts).

    The dataset has the form:
    enron_data["LASTNAME FIRSTNAME MIDDLEINITIAL"] = { features_dict }

    {features_dict} is a dictionary of features associated with that person.
    You should explore features_dict as part of the mini-project,
    but here's an example to get you started:

    enron_data["SKILLING JEFFREY K"]["bonus"] = 5600000
    
"""

import pickle

enron_data = pickle.load(open("../final_project/final_project_dataset.pkl", "rb"))

lenght_enron_data=len(enron_data)
print "number of persons in the dataset: ", lenght_enron_data

#enron_data
import pandas as pd
unpickled_df = pd.read_pickle("../final_project/final_project_dataset.pkl")
#print "Data from enron complete:", unpickled_df

from fractions import Fraction
from __future__ import division

POI_count=0
people_count=0
Salary_quant=0
Email_quant=0
NAN_payments=0
POI_NAN_payments=0

for person in enron_data:
    people_count=people_count+1
    if enron_data[person]['poi']==1:
        POI_count=POI_count+1
        if enron_data[person]['total_payments']=='NaN':
            POI_NAN_payments=POI_NAN_payments+1
    if enron_data[person]['salary']!='NaN':
        Salary_quant=Salary_quant+1
    if enron_data[person]['email_address']!='NaN':
        Email_quant=Email_quant+1    
    if enron_data[person]['total_payments']=='NaN':
        NAN_payments=NAN_payments+1
        
percentageNANpay=(NAN_payments/people_count)*100
percentagePOINANpay=(POI_NAN_payments/POI_count)*100
    
        
print "POI count: ",POI_count

Prentice_Stock=enron_data["PRENTICE JAMES"]["total_stock_value"]
print "Prentice James total stock value: ", Prentice_Stock

POInames=open("../final_project/poi_names.txt")
Wesley_mail2poi=enron_data["COLWELL WESLEY"]["from_this_person_to_poi"]
print "Wesley Colwell emails to POI",Wesley_mail2poi

Skilling_stock_ex=enron_data["SKILLING JEFFREY K"]["exercised_stock_options"]
print "Skilling exercised stock",Skilling_stock_ex

Lay_pay=enron_data["LAY KENNETH L"]["total_payments"]
Skilling_pay=enron_data["SKILLING JEFFREY K"]["total_payments"]
Fastow_pay=enron_data["FASTOW ANDREW S"]["total_payments"]

print "Lay: $",Lay_pay,"Skilling: $",Skilling_pay,"Fastow: $",Fastow_pay

print "Folks quantified salary: ",Salary_quant
print "Folks Known Email: ", Email_quant

print "Folks without total payments:", NAN_payments
print "percentage of Folks with NaN payments:",percentageNANpay,"%"

print "POI Folks without Total Payments: ",POI_NAN_payments
print "percentage POI NaN payments: ",percentagePOINANpay,"%"
```
